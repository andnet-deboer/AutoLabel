```python
import yaml

yaml_path = '/content/yolo_obb_dataset/dataset.yaml'

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Update path to Colab environment
data['path'] = '/content/yolo_obb_dataset'
data['train'] = 'images/train'
data['val'] = 'images/val'

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)
```

```python
# Load model
model = YOLO('yolov8n-obb.pt')

# Train
results = model.train(
    data='/content/yolo_obb_dataset/dataset.yaml',
    epochs=30,
    imgsz=640,
    batch=32,
    device=0,
    name='train_cars_model'
)
```

```python
model = YOLO('yolov8n-obb.pt')

results = model.train(
    data='/content/yolo_obb_dataset/dataset.yaml',
    epochs=60,
    imgsz=640,
    batch=16,
    device=0,
    name='augmented_model',
    mosaic=1.0,
    copy_paste=0.4,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2,
)
```
